### Import Player Salary Information

```python
nbasalary = "playersalary.csv"
nbasalary_df = pd.read_csv(nbasalary)
nbasalary_df.head()
```

### Tranform NBA Player Salary Dataframe and Rename Columns

```python
nba1920_df = nbasalary_df[["Player", "2019-20"]].copy()
nba1920_df.head()
```

```python
# Renamed columns to lower case format and changed 2019-2020 header to salary for ease of use
salary1920_transformed = nba1920_df.rename(columns={"Player":"player","2019-20": "salary"})
salary1920_transformed.head()
```

### Import NBA Player Stats Dataframe and Rename Columns

```python
nbastats = "playerstats.csv"
nbastats_df = pd.read_csv(nbastats)
nbastats_df
```

```python
nbastats1920_df = nbastats_df[["Player", "GP", "FG%", "PPG"]].copy()
nbastats1920_df.head()
```

```python
# Renamed columns to lower case format for ease of use
nbastats1920_transformed = nbastats1920_df.rename(columns={"Player": "player", "GP": "gp","FG%": "fg_percent","PPG":"ppg"})
nbastats1920_transformed.head()
```

### Create Database Connection

```python
connection_string = "postgres:postgres@localhost:5432/etl_project"
```

```python
enginesql = create_engine(f'postgresql://{connection_string}')
```

```python
#Confirming connection by checking table names
enginesql.table_names()
```

### Use pandas to load csv converted DataFrame into database

```python
nbastats1920_transformed.to_sql(name='player_stats', con =enginesql, if_exists= 'append', index=False)
```

```python
salary1920_transformed.to_sql(name='player_salary', con=enginesql, if_exists='append', index=False)
```

### Confirm data has been added by querying the player_stats table

```python
pd.read_sql_query('select * from player_stats', con=enginesql).head()
```

### Confirm data has been added by querying the player_salary table

```python
pd.read_sql_query('select * from player_salary', con=enginesql).head()
```
